```python
#data['LoanAmount'].fillna(data['LoanAmount'].mean(), inplace=True)
data['Self_Employed'].fillna('No',inplace=True)
```

```python
#create pivot table
table = data.pivot_table(values='LoanAmount', index='Self_Employed' ,columns='Education', aggfunc=np.median)
print(table)
def fage(x): 
    return table.loc[x['Self_Employed'],x['Education']] 
# Replace missing values
data['LoanAmount'][data['LoanAmount'].isnull()] = data[data['LoanAmount'].isnull()].apply(fage, axis=1)
```

```python
data['LoanAmount_log'] = np.log(data['LoanAmount'])
data['LoanAmount_log'].hist(bins=20)
```

```python
#Features created
data['TotalIncome'] = data['ApplicantIncome'] + data['CoapplicantIncome']
data['TotalIncome_log'] = np.log(data['TotalIncome'])
data['LoanAmount_log'].hist(bins=20)
```

```python
data.head(100)
```

```python
#impute missing values with most frequent value
data['Gender'] = data['Gender'].fillna(data['Gender'].value_counts().index[0])
data['Married'] = data['Married'].fillna(data['Married'].value_counts().index[0])
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].value_counts().index[0])
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].value_counts().index[0])
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
```

```python
data.apply(lambda x: sum(x.isnull()),axis=0)
#(data['Gender'].value_counts().index[0])
```

```python
data['Dependents'].value_counts()
data['is_edu'] = data['Education'].apply(lambda x: 1 if x == 'Graduate' else 0)
data['is_Nonedu'] = data['Education'].apply(lambda x: 1 if x == 'Not Graduate' else 0)
data['dep0'] = data['Dependents'].apply(lambda x: 0 if x ==0  else 0)
data['dep1'] = data['Dependents'].apply(lambda x: 1 if x ==1  else 0)
data['dep2'] = data['Dependents'].apply(lambda x: 2 if x ==2  else 0)
data['dep3'] = data['Dependents'].apply(lambda x: 3 if x ==3  else 0)
```

```python
#print(data['Dependents'].value_counts())
data["Total_Income"] = data.CoapplicantIncome + data.ApplicantIncome
data['Income_ratio'] = data.Total_Income / data.LoanAmount
```

```python
data.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
#Convertin Data into int
from sklearn.preprocessing import LabelEncoder
labelEnc = LabelEncoder()

data['Gender'] = labelEnc.fit_transform(data['Gender'])
data['Loan_Status'] = labelEnc.fit_transform(data['Loan_Status'])
data['Education'] = labelEnc.fit_transform(data['Education'])
data['Self_Employed'] = labelEnc.fit_transform(data['Self_Employed'])
data['Property_Area'] = labelEnc.fit_transform(data['Property_Area'])
data['Married'] = labelEnc.fit_transform(data['Married'])
#data['Dependents'] = labelEnc.fit_transform(data['Dependents'])

data['is_male'] = data['Gender'].apply(lambda x: 1 if x ==1  else 0)
data['is_female'] = data['Gender'].apply(lambda x: 1 if x ==0  else 0)

#data["Gender"] = data["Gender"].astype('category')
```

```python
data.head()
```

```python
# Splitting Data
from sklearn.model_selection import train_test_split

#features_to_use = ['Gender','Married','Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome'
     #              , 'Credit_History', 'Property_Area', 'LoanAmount_log', 'TotalIncome_log', 'Income_ratio']
#features_to_use = ['is_male','is_female','Married','Dependents', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'Credit_History', 'Property_Area', 'LoanAmount_log', 'TotalIncome_log', 'Income_ratio']
#features_to_use = ['dep0','dep1', 'dep2','dep3','is_male','is_female','Married', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'Credit_History', 'Property_Area', 'LoanAmount_log', 'TotalIncome_log', 'Income_ratio']   
#features_to_use = ['Married', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'Credit_History', 'Property_Area', 'LoanAmount_log', 'TotalIncome_log', 'Income_ratio']   
features_to_use = ['is_edu', 'is_Nonedu','dep0','dep1', 'dep2','dep3','is_male','is_female','Married', 'Education', 'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'Credit_History', 'Property_Area', 'LoanAmount_log', 'TotalIncome_log', 'Income_ratio']   

X = data[features_to_use]
y = data['Loan_Status']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2)
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
#from xgboost import XGBRegressor
```

```python
#creating Model
my_model = XGBClassifier(n_estimators=250, learning_rate=0.05)
```

```python
#from sklearn import linear_model
#reg = linear_model.LogisticRegression() 
#reg.fit(X, y) 
```

```python
#Training Model
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)],
             verbose=False)
```

```python
#Testing on validation data
my_predictions = my_model.predict(X_valid)
from sklearn.metrics import accuracy_score,f1_score
accuracy = accuracy_score(my_predictions, y_valid)
print(accuracy)
```
