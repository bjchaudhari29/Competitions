```python
out = []
for j in range(len(data_test)):    
#for j in range(0,4):
# test query
    p = np.array(data_test.ix[j])[1]
    q = np.array(data_test.ix[j])[2:p+2]
    dict = {}
    for i in range(len(data_train)):
        # train data
        a = np.array(data_train.ix[i])[1]
        b = np.array(data_train.ix[i])[2:a+2]
        x,y = (q, find_closest(q, b))
#        print()
        #print('\n')
        dict[data_train.file_name[i]] = np.linalg.norm(x-y)
    #print(j)
    temp = min(dict.values())
    res = [key for key in dict if dict[key] == temp]
   # print(temp, res)
    out.append(res)
```
